```python

    # ------------ Main loop --------------
counter = 0     
for f in tqdm(range(len(files)),desc='Images checked'):
            
    file_path=files[f]
    try:
      img = np.array(io.imread(file_path))
      if np.amax(img[0,:,:]) == 0:
        os.remove(file_path)
        counter +=1
        display('Image id: '+ files[f] + ' was removed. Total images found: ' + str(counter))
    except:
      os.remove(file_path)
      counter +=1
      display('Image id: '+ files[f] + ' was removed. Total images found: ' + str(counter))
            
print('Number of images removed:')
print('counter')
```
